```python
import tkinter as tk
from PIL import ImageTk, Image

class GUIWindow:
    def __init__(self, root):
        self.root = root
        self.root.title("발음 교정 피드백")
        self.root.geometry("400x400")
        # 이미지 파일 경로 설정
        self.vowel = ['ㅏ.png', 'ㅓㅗ.png', 'ㅔ.png', 'ㅡㅜ반모음W.png', 'ㅣ반모음J.png', 'AA-ㅏ.png', 'EE-ㅔ.png', 'EO-ㅓ.png', 'eu-ㅡ.png', 'ii-l.png', 'JY-반모음J.png', 'OO-ㅗ.png', 'UU-ㅜ.png', 'WW-반모음W.png']
        self.consonant = ['ㅍPM-ㅁㅂㅃㅍ-혀.png', 'J-ㅈㅉㅊ.png', 'KNG-ㄱㄲㅋㅇ.png', 'L-ㄹ.png', '', 'S-ㅅㅆ.png', 'SH-변이음ㅅㅆ.png', 'TN-ㄴㄷㄸㅌ.png']
        self.add = ['ㅍPM-ㅁㅂㅃㅍ-혀.png', 'Rest.png']

        self.feedback_dict = {'ㅏ': self.vowel[0], 'ㅓ': self.vowel[1]}

        # 이미지 및 설명을 표시할 라벨 생성
        self.image_label = tk.Label(self.root, padx=10, pady=10)
        self.description_label = tk.Label(self.root)
        self.image_label.pack()
        self.description_label.pack()

        # 입력 받는 창과 버튼 생성
        self.input_entry = tk.Entry(self.root)
        self.input_entry.pack()
        self.submit_button = tk.Button(self.root, text="입력", command=self.process_input)
        self.submit_button.pack()

        self.current_index = -1
        self.feedback_values = []
        self.next_button = tk.Button(self.root, text="다음", command=self.show_next_image, state='disabled')
        self.next_button.pack()
        self.close_button = tk.Button(self.root, text="닫기", command=self.root.destroy, state='disabled')
        self.close_button.pack()

    def show_image_and_description(self):
        if self.current_index >= len(self.feedback_values):
            self.image_label.configure(text="완벽합니다!")
            self.description_label.configure(text="")
            self.input_entry.pack_forget()
            self.submit_button.pack_forget()
            self.next_button.configure(state='disabled')
            self.close_button.configure(state='normal')
        else:
            value = self.feedback_values[self.current_index]
            if value in self.feedback_dict:
                image_path = self.feedback_dict[value]
                image = Image.open(image_path)
                image = image.resize((300, 300))
                photo = ImageTk.PhotoImage(image)
                self.image_label.configure(image=photo)
                self.image_label.image = photo
                self.description_label.configure(text=f"Image {self.current_index+1} description")
                if self.current_index < len(self.feedback_values) - 1:
                    self.next_button.configure(state='normal')
                else:
                    self.next_button.configure(state='disabled')
                    self.close_button.configure(state='normal')
            else:
                self.image_label.configure(text="완벽합니다!")
                self.description_label.configure(text="")
                self.input_entry.pack_forget()
                self.submit_button.pack_forget()
                self.next_button.configure(state='disabled')
                self.close_button.configure(state='normal')

    def process_input(self):
        feedback = self.input_entry.get()
        self.input_entry.delete(0, 'end')
        self.feedback_values = [f.strip() for f in feedback.split(',')]
        self.current_index = 0
        self.input_entry.pack_forget()
        self.submit_button.pack_forget()
        self.show_image_and_description()

    def show_next_image(self):
        self.current_index += 1
        self.show_image_and_description()

root = tk.Tk()
gui_window = GUIWindow(root)
root.mainloop()
```
